```python
import numpy as np
import tensorflow as tf
from model import PointerNet
from data import Loader, tokenize
```

```python
tf.reset_default_graph()
m = PointerNet()

save_dir = './models/'
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
saver = tf.train.Saver(tf.global_variables())
ckpt = tf.train.get_checkpoint_state(save_dir)
saver.restore(sess, ckpt.model_checkpoint_path)
```

```python
testing = Loader('./data/test.txt', './data/vocab.json', 100, 45)
```

```python
x_batch, x_lengths, _ = testing.next_batch()
```

```python
pointers = sess.run(m.pointers, feed_dict={m.encoder_inputs: x_batch,
                                           m.input_lengths: x_lengths})
pointers = pointers[0]
```

```python
for i in range(100):
    text = testing.text[i]
    print(text)
    print('Answer: {} \n'.format(tokenize(text)[pointers[i]]))
```
